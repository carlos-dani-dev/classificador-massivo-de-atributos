```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import BorderlineSMOTE
from cuml.neighbors import NearestNeighbors
import numpy as np
```

```python
print( df["fitz_type"].values[0] )
print( df["Face_embedding"].values[0].shape )
```

```python
X = np.vstack(df["Face_embedding"].values)
y = df["fitz_type"].values

nn = NearestNeighbors(n_neighbors = 6)

X_resampled, y_resampled = SMOTE(k_neighbors=nn).fit_resample(X, y)
```

```python
df_resampled = pd.DataFrame({"face embedding": list(X_resampled), "fitz_type": y_resampled})
df_resampled
```

```python
df_resampled.to_pickle("fitz_type_balanced_embeddings_casual.pkl")
```
