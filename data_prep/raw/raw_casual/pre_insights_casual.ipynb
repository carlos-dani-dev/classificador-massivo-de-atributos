```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
anno = pd.read_json("CasualConversationsV2.json")
anno.columns
```

```python
age_classes = {
    "young adult", # 18 à 29
    "adult", # 30 à 64
    "elderly" # 65+
}
bins = [18, 30, 65, 150]


def get_type(value):
    print(value)
    return value


def get_scale(value):
    print(value)
    return value
```

```python
anno_filtered = anno[["video_name", "subject_id", "age", "fitzpatrick_skin_tone", "monk_skin_tone"]]
anno_filtered = anno_filtered[anno_filtered["fitzpatrick_skin_tone"].apply(
    lambda x : x.get('confidence') != "low"
)]
anno_filtered["fitzpatrick_skin_tone"] = anno_filtered["fitzpatrick_skin_tone"].apply(lambda x : x.get('type'))

anno_filtered = anno_filtered[anno_filtered["monk_skin_tone"].apply(
    lambda x : x.get('confidence') != "low"
)]
anno_filtered["monk_skin_tone"] = anno_filtered["monk_skin_tone"].apply(lambda x : x.get('scale'))

anno_filtered = anno_filtered[anno_filtered["age"].apply(
    lambda x : x != "prefer not to say"
)]

anno_filtered["age"] = pd.cut(anno_filtered["age"], bins=bins, labels=age_classes)
```
